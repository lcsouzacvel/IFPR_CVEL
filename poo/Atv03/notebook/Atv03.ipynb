Mostre a média usando o showMessageDialog da classe JOptionPane

```java
import javax.swing.JOptionPane;
String aux;
float p1=0,p2=0,t1=0,media=0;
try{ 
aux=JOptionPane.showInputDialog("Digite a primeira nota");
p1=Float.parseFloat(aux);
aux=JOptionPane.showInputDialog("Digite a primeira nota");
p2=Float.parseFloat(aux);
aux=JOptionPane.showInputDialog("Digite a primeira nota");
t1=Float.parseFloat(aux);
}
catch(NumberFormatException erro){
JOptionPane.showMessageDialog(null,"Digite somente numeros" + erro.toString());
}
media =((p1 +p2 +t1)/3);
// mostrando a mensagem na tela com JOptionpane
JOptionPane.showMessageDialog(null,"Nota 1:"+ p1 +"Nota 2:"+ p2 +"Trablho:"+ t1 +"Média notas foi="+ media);
```
